# Question: 
## Having a porch increases home sale price

Null hypothesis
H_0: Having a porch does not increase home sale price.

Alternative hypothesis
H_a: Having a porch does increase home sale price.

```python
df_res_bldg_test = df_res_bldg[['Major', 'Minor', 'SqFtOpenPorch', 'SqFtEnclosedPorch']]
df_res_bldg_test_open = df_res_bldg_test[df_res_bldg_test['SqFtOpenPorch'] > 0] 
df_res_bldg_test_enclosed = df_res_bldg_test[df_res_bldg_test['SqFtEnclosedPorch'] > 0]

# create a new column for the total square footage of porch
df_res_bldg_test['Porch'] = df_res_bldg_test['SqFtOpenPorch'] + df_res_bldg_test['SqFtEnclosedPorch']

# check - display 10 random results where the sqare footage is > 0
df_res_bldg_test[df_res_bldg_test['Porch'] > 0].sample(10)
```

```python
df_property_sale_model_test = df_property_sale_type[['Major', 'Minor', 'SalePrice']]
df_property_sale_model_test.sample(5)
```

```python
df_test = pd.concat([df_property_sale_model_test, df_res_bldg_test], axis=1, join='inner')
df_test[df_test['Porch'] > 0].sample(10)
```

```python
conditions = [
    (df_test['Porch'] > 0),
    (df_test['Porch'] == 0)]
choices = [1, 0]

df_test['Has_Porch'] = np.select(conditions, choices, default='null')


df_test.sample(10)
```

```python
df_yes = df_test[df_test['Has_Porch'] == '1']
print(df_yes.info())
df_yes['SalePrice'].mean()
```

```python
df_no = df_test[df_test['Has_Porch'] == '0']
print(df_no.info())
df_no['SalePrice'].mean()
```
